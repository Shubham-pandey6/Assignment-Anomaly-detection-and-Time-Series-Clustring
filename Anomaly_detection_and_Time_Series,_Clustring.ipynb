### Question 1: What is Dimensionality Reduction? Why is it important in machine learning?  
**Answer:**  
Dimensionality Reduction is the process of reducing the number of input variables (features) in a dataset while preserving as much relevant information as possible.  
- It helps simplify complex datasets.  
- Reduces computational cost and training time.  
- Minimizes overfitting by removing redundant/noisy features.  
- Improves visualization of high-dimensional data.  

---

### Question 2: Name and briefly describe three common dimensionality reduction techniques.  
**Answer:**  
- **Principal Component Analysis (PCA):** Transforms data into new axes (principal components) that maximize variance.  
- **Linear Discriminant Analysis (LDA):** Focuses on maximizing class separability by projecting data onto lower dimensions.  
- **t-Distributed Stochastic Neighbor Embedding (t-SNE):** Non-linear technique that preserves local similarities, often used for visualization of high-dimensional data.

### Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.  
**Answer:**  
Clustering is an unsupervised learning technique that groups similar data points together based on their features without predefined labels.  
- **K-Means Clustering:** Divides data into k clusters by minimizing intra-cluster variance.  
- **Hierarchical Clustering:** Builds a tree-like structure of clusters using agglomerative or divisive methods.  
- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):** Groups points based on density and identifies outliers as noise.

### Question 4: Explain the concept of anomaly detection and its significance.  
**Answer:**  
Anomaly detection is the process of identifying unusual patterns or data points that deviate significantly from the norm.  
- **Significance:**  
  - Detects fraud in financial transactions.  
  - Identifies faults in industrial systems.  
  - Improves cybersecurity by spotting intrusions.  
  - Ensures data quality by removing outliers.  

---

### Question 5: List and briefly describe three types of anomaly detection techniques.  
**Answer:**  
- **Statistical Methods:** Use probability distributions to detect data points that lie outside expected ranges.  
- **Machine Learning Models:** Algorithms like Isolation Forest or Autoencoders learn normal patterns and flag deviations.  
- **Distance/Clustering-Based Methods:** Outliers are detected based on their distance from cluster centers or nearest neighbors.

### Question 6: What is time series analysis? Mention two key components of time series data.  
**Answer:**  
Time series analysis involves studying data points collected or recorded at successive time intervals to identify trends, patterns, and seasonal effects.  
- **Key Components:**  
  - **Trend:** Long-term increase or decrease in data values.  
  - **Seasonality:** Regular, repeating patterns over fixed periods (e.g., monthly sales spikes).  

---

### Question 7: Describe the difference between seasonality and cyclic behavior in time series.  
**Answer:**  
- **Seasonality:** Refers to predictable, fixed, and repeating patterns (e.g., holiday sales every December).  
- **Cyclic Behavior:** Refers to irregular fluctuations influenced by economic or social factors, without a fixed period (e.g., business cycles).

### Question 10: How can time series analysis be used for anomaly detection?  
**Answer:**  
Time series analysis can detect anomalies by identifying deviations from expected temporal patterns.  
- **Trend-based anomalies:** Sudden spikes or drops in data.  
- **Seasonality-based anomalies:** Irregular behavior during expected seasonal cycles.  
- **Residual analysis:** Comparing observed values with predicted values from forecasting models (e.g., ARIMA, LSTM).  

````python
'''### Question 8: Write Python code to perform K-means clustering on a sample dataset.
**Answer:**
'''
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.show()


'''**Output:**
A scatter plot showing data points grouped into 4 clusters, with red dots marking cluster centers.

'''
````
